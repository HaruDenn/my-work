```python
#nomor 2

imgd = cv2.imread('doggo.jpg')
imgd
```

```python
plt.imshow(imgd)
```

```python
imgs = imgd.copy()
points = np.array([[30,58],[25,65],[37,80],[50,65]],np.int32)
imgdZ = cv2.putText(imgs,text='I',org=(35,20),fontFace = cv2.FONT_HERSHEY_COMPLEX,fontScale = -1,color= (255,255,255),thickness= 1, lineType=cv2.LINE_AA)
imgdZ = cv2.circle(imgs, center=(30,60), radius=10, color=(255,0,0), thickness=-1)
imgdZ = cv2.circle(imgs, center=(45,60), radius=10, color=(255,0,0), thickness=-1)
imgdZ = cv2.fillPoly(imgs,[points],(255,0,0))
imgdZ = cv2.putText(imgs,text='Dog',org=(10,100),fontFace = cv2.FONT_HERSHEY_COMPLEX,fontScale = 1,color= (255,255,255),thickness= 1, lineType=cv2.LINE_AA)
plt.imshow(imgdZ)
```
